```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
danger = [9, 8, 90, 85, 80, 70, 70, 65, 55, 60, 70, 65, 50]
railgun = [75, 70, 60, 65, 60, 45, 55, 50, 40, 40, 35, 35, 30]

time = np.arange(0, 130, 10)
```

```python
len(danger), len(railgun)
```

```python
plt_danger, = plt.plot(time, danger, color='red', label='Danger Drop')
plt_railgun, = plt.plot(time, railgun, color='blue', label='Railgun')
plt.xlabel('Time in Seconds')
plt.ylabel('Speed in MPH')
plt.title('Coaster Speed Over Runtime')
plt.xlim(0, 125)
plt.ylim(5, 100)
plt.legend(handles=[plt_danger, plt_railgun],loc='best')
plt.grid()
```
