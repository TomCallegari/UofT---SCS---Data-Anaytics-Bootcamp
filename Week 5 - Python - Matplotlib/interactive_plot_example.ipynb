```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_axis = np.arange(0, 10, 0.1)
sin = np.sin(x_axis)
cos = np.cos(x_axis)
```

```python
plt.hlines(0, 0, 10, alpha=0.25)
```

```python
plt.plot(x_axis, sin, marker = 'o', color='blue', label='Sine')
plt.plot(x_axis, cos, marker='^', color='red', label='Cos')
```

```python
plt.legend(loc='lower right')
```

```python
plt.savefig('lineConfig.png')
```

```python
plt.show()
```
