```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_axis = np.arange(0, 5, 0.1)
x_axis
```

```python
e_x = [np.exp(x) for x in x_axis]
e_x
```

```python
plt.plot(x_axis, e_x)
plt.xlabel('Time')
plt.ylabel('How cool matplotlib is')
```

```python
x_axis = np.arange(0, 6, 0.1)
sin = np.sin(x_axis)
cos = np.cos(x_axis)
```

```python
plt.plot(x_axis, sin)
plt.plot(x_axis, cos)
```
