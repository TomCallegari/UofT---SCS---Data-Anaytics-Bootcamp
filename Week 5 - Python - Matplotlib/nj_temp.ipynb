```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
temp = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
months = np.arange(1, 13, 1)
```

```python
cel = [(i - 32)*0.56 for i in temp]
```

```python
plt.plot(months, temp, marker='+', color='blue', label='Fahrenheit')
plt.plot(months, cel, marker='s', color='red', label='Celsius')
plt.legend(loc='best')
plt.xlabel('Months')
plt.ylabel('Temperature in Degrees')
```

```python
plt.plot(months, temp)
plt.plot(months, cel)
```
